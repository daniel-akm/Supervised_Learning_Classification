```python
import pandas as pd
df = pd.read_csv('Data/New_Driver_Application_Status.csv')
df
```

```python
X = df.iloc[:,df.columns.isin(['Type', 'Drug Test', 'WAV Course', 'Defensive Driving', 'Driver Exam', 'Medical Clearance Form'])]
X.head()
```

```python
y = df.iloc[:, df.columns.isin(['Status'])]
X['Drug Test'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
X['Type'] = lb.fit_transform(X['Type'])
X['Drug Test'] = lb.fit_transform(X['Drug Test'])
X['WAV Course'] = lb.fit_transform(X['WAV Course'])
X['Defensive Driving'] = lb.fit_transform(X['Defensive Driving'])
X['Driver Exam'] = lb.fit_transform(X['Driver Exam'])
X['Medical Clearance Form'] = lb.fit_transform(X['Medical Clearance Form'])
y['Status'] = lb.fit_transform(y['Status'])
X['Drug Test'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,shuffle=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier(n_neighbors=3)
knc.fit(X_train,y_train)
```

```python
pred_values = knc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
print(f'Accuracy score is {np.round(accuracy_score(y_test,pred_values)*100,2)}%')
```

```python
import pickle as pkl
with open('project1.pkl','wb') as f:
    pkl.dump(knc, f)
```
